# Exploring Montreal Transect Data Set

```python
import pandas as pd

df = pd.read_excel (r'data/Montreal-Transect-Data-2022.xlsx')
```

```python
# Make prettier, more interactive plots
import pandas_bokeh
pandas_bokeh.output_notebook()
```

```python
df.columns
```

```python
# Manage things with `Metro Station` instead of `Station Number`.
df = df.drop(columns='Station Number')
```

```python
df.plot_bokeh(x='Metro Station', vertical_xlabel=True,figsize=(800,600))
```

Observations after exploring the graph. I turned on an off variables and looked for things that moved together or opposite of each other. 

```python
df.plot_bokeh(x='Metro Station', y = ['Language','Building Height'], vertical_xlabel=True,
              figsize=(600,400))
```

```python
df.plot_bokeh(x='Metro Station', y = ['Language','Land Use - Residential'], vertical_xlabel=True,
              figsize=(600,400))
```

```python
df.plot_bokeh(x='Metro Station', y = ['Investment-Disinvestment','Land Use - Residential'], vertical_xlabel=True,
              figsize=(600,500))
```
